```python
import os, matplotlib.pyplot as plt
import cv2,numpy as np
```

```python
face = cv2.CascadeClassifier('C:\\Users\\ASUS\\AppData\\Local\\Programs\\Python\\Python38\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('F:/Code_Python/ThiGiacMayTinh/ComputerVision_IUH/CV_Project/GAN_Project/Computer_vision/img/5.png')[:,:,::-1]
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
imgs = []
for i in os.listdir('img/'):
  img = cv2.imread('F:/Code_Python/ThiGiacMayTinh/ComputerVision_IUH/CV_Project/GAN_Project/Computer_vision/img/'+ i)[:,:,::-1]
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  img = np.array(img)
  faces = face.detectMultiScale(gray,1.1,4)
  for x,y,w,h in faces:
    gray_ = gray[y-30:y+h+30 , x-30:x+w+30]
    color = img[y-30:y+h+30 , x-30:x+w+30]
    facess = face.detectMultiScale(gray_)
    # plt.imshow(color)
    # plt.show()
    if len(facess) == 0:
      print('False')
    else:
      for (ex,ey,ew,eh) in facess:
        face_roi = color
        face_roi = cv2.cvtColor(face_roi,cv2.COLOR_BGR2RGB)
        imgs.append(face_roi)
```

```python
for i in range(len(imgs)):
    cv2.imwrite(f'im/{i}.png',imgs[i])
```
